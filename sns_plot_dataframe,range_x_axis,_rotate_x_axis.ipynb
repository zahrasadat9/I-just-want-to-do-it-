```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

a= pd.DataFrame({'Name': ['Tom', 'Joseph', 'Krish', 'John'], 'Age': [20, 18, 19, 18]}   )
# blue grid lined BG
sns.set_theme()
# fig size
plt.figure(figsize=(15,8))

ax= sns.histplot(a,kde=True,bins=4)
# range of plot
ax.set_xticks(range(15,25));
# rotate the x axis labels without changing hem
ax.tick_params(axis='x', labelrotation=90);
```
